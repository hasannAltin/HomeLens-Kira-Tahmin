```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")


df = pd.read_csv("homelens_temsili_veri.csv")

print(df.head())

print(df.info())

print(df.describe())
```

```python
sayisal_sutunlar = ['Kira', 'Metrekare', 'Bina_Yasi', 'Aidat']

plt.figure(figsize=(15, 8))

for i, sutun in enumerate(sayisal_sutunlar):
    plt.subplot(2, 2, i + 1)
    sns.histplot(df[sutun], kde=True, color='skyblue')
    plt.title(f"{sutun} Dağılımı")
    plt.xlabel(sutun)
    plt.ylabel("Frekans")

plt.tight_layout()
plt.show()
```

Semtlere Göre Ortalama Kira

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Semt', y='Kira', estimator='mean', errorbar=None, palette='viridis')
plt.title("Semtlere Göre Ortalama Kira")
plt.ylabel("Ortalama Kira (TL)")
plt.xlabel("Semt")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

M² İle Kira Arası İlişki


```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Metrekare', y='Kira', hue='Semt', palette='tab10')
plt.title("Metrekare vs Kira (Semtlere Göre)")
plt.xlabel("Metrekare")
plt.ylabel("Kira")
plt.legend(title='Semt')
plt.tight_layout()
plt.show()
```

Korelasyon Matrisi (Isı Haritası)

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df[['Kira', 'Metrekare', 'Bina_Yasi', 'Aidat']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Sayısal Değişkenler Arası Korelasyon")
plt.tight_layout()
plt.show()
```
